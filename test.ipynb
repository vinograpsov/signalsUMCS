```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
def load_signals(signal1, signal2):
    df1 = pd.read_csv(signal1, sep=';', header=None)
    df2 = pd.read_csv(signal2, sep=';', header=None)
    return df1, df2

def convolve_signalse(df1, df2):
    signal1 = df1.iloc[:,0]
    signal2 = df2.iloc[:,0]
    result = np.convolve(signal1, signal2, mode='full')
    return result
def plot_signals(df1, df2, result):
    plt.figure()
    plt.plot(df1.iloc[:,0])
    plt.title("signal1")
    plt.show()

    plt.figure()
    plt.plot(df2.iloc[:,0])
    plt.title("signal2")
    plt.show()

    plt.figure()
    plt.plot(result)
    plt.plot(result)
    plt.title("result")
    plt.show()
```

```python
# file1 = 'sin1_1_1000.csv'
# file2 = 'sin3_1_1000.csv'

# file1 = 'los_gauss_1_1000.csv'
# file2 = 'los2_1_1000.csv'

file1 = 'syg3_1_256.csv'
file2 = 'syg6_1_256.csv'

df1, df2 = load_signals(file1, file2)
result = convolve_signalse(df1, df2)
plot_signals(df1, df2, result)
```
